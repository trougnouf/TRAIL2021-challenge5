```python
import torch
import numpy as np
import torch.nn as nn
from torchvision.datasets import ImageFolder
from torch.utils.data import DataLoader
from torchvision.transforms import ToTensor, Resize, Compose, Normalize, CenterCrop
import torchvision
import matplotlib.pyplot as plt
import torchvision.transforms.functional as TF

plt.rcParams["figure.figsize"] = (15,15)

from utils import Augmenter, UnNormalizer
```

```python
train_set = ImageFolder(
    root="~/.cache/torch/mmf/data/pneumonia/train",
    transform=Compose([
            Resize(374),
            CenterCrop(224),
            Augmenter(ra=False, prob=0.5),
            ToTensor(),
            #Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])
)
```

```python
def show(imgs):
    if not isinstance(imgs, list):
        imgs = [imgs]
    fix, axs = plt.subplots(ncols=len(imgs[0]), squeeze=False)
    for i, img in enumerate(imgs[0]):
        img = img.detach()
        img = TF.to_pil_image(img)
        axs[0, i].imshow(np.asarray(img))
        axs[0, i].set(xticklabels=[], yticklabels=[], xticks=[], yticks=[])
```

```python
train = DataLoader(train_set, batch_size=1, shuffle=False)
```

```python
for i, data in enumerate(train):
    inputs, labels = data
    print(labels)
    break
```

```python
grid = torchvision.utils.make_grid([inputs])
show(grid)
```
